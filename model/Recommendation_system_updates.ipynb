```python
from collections import defaultdict
from surprise import accuracy
from surprise.model_selection import KFold

def precision_recall_at_k(predictions, k=10, threshold=3.5):
    '''Return precision and recall at k metrics for each user.'''
    # First map the predictions to each user.
    user_est_true = defaultdict(list)
    for uid, _, true_r, est, _ in predictions:
        user_est_true[uid].append((est, true_r))

    precisions = dict()
    recalls = dict()

    for uid, user_ratings in user_est_true.items():
        # Sort user ratings by estimated value
        user_ratings.sort(key=lambda x: x[0], reverse=True)
        # Number of relevant items
        n_rel = sum((true_r >= threshold) for (_, true_r) in user_ratings)
        # Number of recommended items in top k
        n_rec_k = sum((est >= threshold) for (est, _) in user_ratings[:k])
        # Number of relevant and recommended items in top k
        n_rel_and_rec_k = sum(((true_r >= threshold) and (est >= threshold))
                              for (est, true_r) in user_ratings[:k])

        # Precision@K: Proportion of recommended items that are relevant
        # Recall@K: Proportion of relevant items that are recommended
        precisions[uid] = n_rel_and_rec_k / n_rec_k if n_rec_k != 0 else 1
        recalls[uid] = n_rel_and_rec_k / n_rel if n_rel != 0 else 1

    return precisions, recalls

# Use the precision_recall_at_k function during model evaluation
kf = KFold(n_splits=5)
for trainset, testset in kf.split(data):
    algo.fit(trainset)
    predictions = algo.test(testset)
    precisions, recalls = precision_recall_at_k(predictions, k=5, threshold=4)
    # Precision and recall can then be averaged over all users
    print(f"Precision: {sum(precisions.values()) / len(precisions):.2f}")
    print(f"Recall: {sum(recalls.values()) / len(recalls):.2f}")
```

```python
from surprise import Dataset, Reader, KNNBasic
from surprise.model_selection import train_test_split
from surprise import accuracy

# Setup the data
reader = Reader(rating_scale=(1, 5))
data = Dataset.load_from_df(Salem_df[['id', 'listing_id', 'predicted_rating']], reader)

# Define the configuration sets for similarity options
similarity_options = [
    {'name': 'msd', 'user_based': True, 'min_support': 1},
    {'name': 'msd', 'user_based': True, 'min_support': 5},
    {'name': 'cosine', 'user_based': True, 'min_support': 1},
    {'name': 'cosine', 'user_based': True, 'min_support': 5},
    {'name': 'pearson', 'user_based': True, 'min_support': 1},
    {'name': 'pearson', 'user_based': True, 'min_support': 5},
    {'name': 'pearson_baseline', 'user_based': True, 'min_support': 1},
    {'name': 'pearson_baseline', 'user_based': True, 'min_support': 5},
]

best_rmse, best_mae, best_params = float("inf"), float("inf"), None

# Iterate over each set of similarity options
for sim_options in similarity_options:
    algo = KNNBasic(sim_options=sim_options)
    trainset, testset = train_test_split(data, test_size=0.25)
    algo.fit(trainset)
    predictions = algo.test(testset)
    rmse = accuracy.rmse(predictions)
    mae = accuracy.mae(predictions)
    if rmse < best_rmse and mae < best_mae:
        best_rmse, best_mae, best_params = rmse, mae, sim_options

print(f"Best RMSE: {best_rmse}, Best MAE: {best_mae}, Best Params: {best_params}")
```

## Content-based Filtering

This approach recommends items similar to other items.

```python
df =  pd.read_csv('/content/drive/MyDrive/data/master_CSVs/df_encoded.csv')
df
```

```python
top_10000_rows = df.head(10000)
top_10000_rows
```

```python
import pandas as pd
from sklearn.metrics.pairwise import cosine_similarity
from sklearn.preprocessing import StandardScaler

# Assuming df is your DataFrame name
features = top_10000_rows.drop(columns=['id', 'review_scores_rating'])
features = pd.get_dummies(features, drop_first=True)

# Standardize the features
scaler = StandardScaler()
features_scaled = scaler.fit_transform(features)

# Create a DataFrame for the scaled features, indexed by item 'id'
features_df = pd.DataFrame(features_scaled, index=top_10000_rows['id'], columns=features.columns)
```

```python
# Calculate cosine similarity between items
item_similarity = cosine_similarity(features_df)
item_similarity_df = pd.DataFrame(item_similarity, index=features_df.index, columns=features_df.index)
```

```python
def predict_rating(item_id, top_n=5):
    if item_id not in item_similarity_df.index:
        return None

    # Get the top N most similar items
    sim_scores = item_similarity_df.loc[item_id]
    top_items = sim_scores.sort_values(ascending=False).iloc[1:top_n+1]

    # Calculate the weighted average of the ratings
    weights = top_items.values
    ratings = df.set_index('id').loc[top_items.index, 'review_scores_rating']
    weighted_ratings = weights.dot(ratings) / weights.sum()

    return weighted_ratings

# Test the prediction function for a sample item
sample_id = top_10000_rows['id'].iloc[0]
predicted_rating = predict_rating(sample_id)
print(f"Predicted Rating for item {sample_id}: {predicted_rating}")
```

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error

predicted = [predict_rating(id) for id in top_10000_rows['id']]  # Generating predictions for each item

# Calculate RMSE and MAE
rmse = mean_squared_error(top_10000_rows['review_scores_rating'], predicted, squared=False)
mae = mean_absolute_error(top_10000_rows['review_scores_rating'], predicted)

print(f"RMSE: {rmse}")
print(f"MAE: {mae}")
```

```python
def calculate_precision_recall_f1(actual, predicted, threshold=4):
    # Binarize actual and predicted ratings based on the threshold
    actual_binary = [1 if x >= threshold else 0 for x in actual]
    predicted_binary = [1 if x >= threshold else 0 for x in predicted]

    # Calculate precision, recall, f1-score using binary vectors
    # Skipping detailed calculation here; use sklearn's classification_report for detailed metrics
    from sklearn.metrics import precision_score, recall_score, f1_score
    precision = precision_score(actual_binary, predicted_binary)
    recall = recall_score(actual_binary, predicted_binary)
    f1 = f1_score(actual_binary, predicted_binary)

    return precision, recall, f1

precision, recall, f1 = calculate_precision_recall_f1(top_10000_rows['review_scores_rating'], predicted)
print(f"Precision: {precision}, Recall: {recall}, F1-Score: {f1}")
```

## Item-item

```python
import pandas as pd

# Assuming 'df' is your DataFrame
pivot_table =Salem_df.pivot_table(index='listing_id', columns='id', values='predicted_rating').fillna(0)
```

```python
from sklearn.metrics.pairwise import cosine_similarity
import numpy as np

# Compute the cosine similarity matrix
similarity_matrix = cosine_similarity(pivot_table)
similarity_df = pd.DataFrame(similarity_matrix, index=pivot_table.index, columns=pivot_table.index)
```

```python
# Print shapes and indexes to verify alignment
print("Pivot Table Shape: ", pivot_table.shape)
print("Similarity Matrix Shape: ", similarity_df.shape)
print("Pivot Table Index: ", pivot_table.index.equals(similarity_df.index))
```

```python
def predict_ratings(ratings_matrix, similarity_matrix):
    # Convert DataFrame to numpy array if not already
    ratings_array = ratings_matrix.values if isinstance(ratings_matrix, pd.DataFrame) else ratings_matrix
    similarity_array = similarity_matrix.values if isinstance(similarity_matrix, pd.DataFrame) else similarity_matrix

    # Calculate the dot product of ratings matrix and the similarity matrix
    weighted_sum = np.dot(ratings_array, similarity_array)

    # Compute the sum of similarities for each item
    sum_of_similarities = np.array([np.abs(similarity_array).sum(axis=1)])

    # Normalize the weighted sum by the sum of similarities
    # Reshape sum_of_similarities for broadcasting to align with weighted_sum
    predicted_ratings = weighted_sum / sum_of_similarities.T

    # Return as DataFrame to maintain structure
    return pd.DataFrame(predicted_ratings, index=ratings_matrix.index, columns=ratings_matrix.columns)

# Now try to predict the ratings again
predicted_ratings = predict_ratings(pivot_table, similarity_df)
```

```python
from sklearn.metrics import mean_squared_error

def evaluate_predictions(true_ratings, predicted_ratings):
    # Flatten the matrices to single arrays for comparison
    true_ratings = true_ratings.values.flatten()
    predicted_ratings = predicted_ratings.values.flatten()

    # Filter out zero entries which represent missing ratings
    mask = true_ratings > 0
    true_ratings = true_ratings[mask]
    predicted_ratings = predicted_ratings[mask]

    # Calculate RMSE and MAE
    rmse = np.sqrt(mean_squared_error(true_ratings, predicted_ratings))
    mae = np.mean(np.abs(predicted_ratings - true_ratings))

    return rmse, mae

rmse, mae = evaluate_predictions(pivot_table, predicted_ratings)
print(f'RMSE: {rmse}')
print(f'MAE: {mae}')
```

```python
import pandas as pd
from sklearn.metrics.pairwise import cosine_similarity

# Assuming 'df' is your DataFrame
features = df.drop(['id', 'review_scores_rating'], axis=1)  # Exclude non-numeric and target rating columns

# If you have categorical data, ensure it's transformed to a suitable format.
# Here, it looks like the data is already encoded properly for use.
```

```python
# Calculate the cosine similarity matrix
similarity_matrix = cosine_similarity(features)

# Convert to DataFrame for easier manipulation
similarity_df = pd.DataFrame(similarity_matrix, index=df['id'], columns=df['id'])
```

```python
def get_top_similar_items(similarity_df, item_id, top_n=5):
    # Sort the items by similarity to the given item
    sorted_items = similarity_df[item_id].sort_values(ascending=False)

    # Exclude the item itself and get the top n items
    top_items = sorted_items.iloc[1:top_n+1].index.tolist()
    return top_items

# Apply the function to each item and store the results in a new column
df['similar_items'] = df['id'].apply(lambda x: get_top_similar_items(similarity_df, x))
```
