```python
import pandas as pd
from nltk.sentiment.vader import SentimentIntensityAnalyzer
import nltk

# Download VADER lexicon if not already downloaded
nltk.download('vader_lexicon')

# Assuming 'df' is your DataFrame
# Initialize VADER
sia = SentimentIntensityAnalyzer()

# Define a function to convert sentiment to rating
def sentiment_to_rating(sentiment_score):
    if sentiment_score > 0.05:
        return 5  # Positive sentiment
    elif sentiment_score < -0.05:
        return 1  # Negative sentiment
    else:
        return 3  # Neutral sentiment

# Clean and preprocess the comments
# Here, replace HTML tags, convert to lowercase, etc.
df['comments_cleaned'] = df['comments'].str.replace('<br/>', ' ').str.lower()

# Apply sentiment analysis
df['sentiment_score'] = df['comments_cleaned'].apply(lambda x: sia.polarity_scores(x)['compound'])

# Convert sentiment score to rating
df['predicted_rating'] = df['sentiment_score'].apply(sentiment_to_rating)

# Now you have a 'predicted_rating' that serves as an estimated rating based on sentiment
df
```

```python
df = df[["listing_id", "reviewer_id", "predicted_rating"]]
```

```python
import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.cluster import KMeans


# Group reviews by reviewer_id
#grouped_reviews = df.groupby('reviewer_id')['comments'].apply(' '.join).reset_index()


# Convert reviewer comments into TF-IDF features
tfidf_vectorizer = TfidfVectorizer(max_features=1000)  # Adjust max_features as needed
X_tfidf = tfidf_vectorizer.fit_transform(df['comments'])

# Apply KMeans clustering
kmeans = KMeans(n_clusters=5, random_state=42)  # Assuming there are 5 rating levels
kmeans.fit(X_tfidf)

# Assign ratings based on clusters
cluster_ratings = kmeans.labels_ + 1  # Add 1 to start ratings from 1 instead of 0

# Add predicted ratings to DataFrame
df['predicted_rating'] = cluster_ratings

# Display DataFrame with predicted ratings
print(df[['reviewer_id', 'predicted_rating']])
```

```python
# Define chunk size
chunk_size = 30000

# List to store chunks with predicted ratings
chunk_list = []

# Loop through chunks of data
for i in range(0, len(reviews), chunk_size):
    # Select the next chunk of data
    chunk = reviews.iloc[i:i+chunk_size]

    # Drop any rows with missing values in the 'comments' column
    chunk = chunk.dropna(subset=['comments'])

    if not chunk.empty:
        # Convert reviewer comments into TF-IDF features
        tfidf_vectorizer = TfidfVectorizer(max_features=1000)  # Adjust max_features as needed
        X_tfidf = tfidf_vectorizer.fit_transform(chunk['comments'])

        # Apply KMeans clustering
        kmeans = KMeans(n_clusters=5, random_state=42)  # Assuming there are 5 rating levels
        kmeans.fit(X_tfidf)

        # Assign ratings based on clusters
        cluster_ratings = kmeans.labels_ + 1  # Add 1 to start ratings from 1 instead of 0

        # Add predicted ratings to DataFrame
        chunk['predicted_rating'] = cluster_ratings

        # Append chunk to the list
        chunk_list.append(chunk)

# Concatenate chunks into a single DataFrame
df_with_predicted_ratings = pd.concat(chunk_list)

# Merge DataFrame with predicted ratings with the original DataFrame based on a common identifier
merged_df = pd.merge(df, df_with_predicted_ratings[['predicted_rating']], left_index=True, right_index=True)

# Save merged DataFrame back to a CSV file
merged_df.to_csv("/content/drive/MyDrive/data/master_CSVs/predicted_ratings_Boston.csv", index=False)
```

```python
import pandas as pd

# Assuming df is your DataFrame with the necessary data
# Calculating the mean predicted rating for each listing_id
mean_predicted = df.groupby('listing_id')['predicted_rating'].mean().reset_index()
mean_predicted.rename(columns={'predicted_rating': 'mean_predicted_rating'}, inplace=True)

# Assuming hotel_df holds unique listing_id and review_scores_rating
# Merging the mean predicted rating with the review_scores_rating
comparison_df = mean_predicted.merge(hotel_df[['id', 'review_scores_rating']], left_on='listing_id', right_on='id', how='left')

# Drop the extra 'id' column post merge
comparison_df.drop('id', axis=1, inplace=True)

# Create a column to compare mean_predicted_rating with review_scores_rating
comparison_df['rating_difference'] = comparison_df['review_scores_rating'] - comparison_df['mean_predicted_rating']

print(comparison_df)

```

```python
comparison_df = comparison_df[['listing_id','mean_predicted_rating','review_scores_rating','rating_difference' ]]
cleaned_df = comparison_df.dropna()
cleaned_df
```

```python
import pandas as pd
import numpy as np

# Assuming cleaned_df is the DataFrame with no NaN values and contains 'mean_predicted_rating' and 'review_scores_rating'
# Calculating RMSE
rmse = np.sqrt(((cleaned_df['mean_predicted_rating'] - cleaned_df['review_scores_rating']) ** 2).mean())
print(f"Root Mean Square Error (RMSE): {rmse}")

# Calculating Mean Absolute Error (MAE)
mae = np.abs(cleaned_df['mean_predicted_rating'] - cleaned_df['review_scores_rating']).mean()
print(f"Mean Absolute Error (MAE): {mae}")

# Calculating R-squared
ss_res = ((cleaned_df['review_scores_rating'] - cleaned_df['mean_predicted_rating']) ** 2).sum()
ss_tot = ((cleaned_df['review_scores_rating'] - cleaned_df['review_scores_rating'].mean()) ** 2).sum()
r_squared = 1 - (ss_res / ss_tot)
print(f"R-squared: {r_squared}")
```
