```python
import pandas as pd

# Replace 'path_to_your_listings_data.csv' with the path to your dataset
data_path = '/content/drive/MyDrive/DS5460 - Big Data/data/master_CSVs/master_listings.csv'
df = pd.read_csv(data_path)

# Assuming your CSV has columns named 'latitude' and 'longitude'
print(df.head())  # Just to check the data
```

```python
import folium

# Create a map centered around the center of USA
map = folium.Map(location=[39.8283, -98.5795], zoom_start=4)  # Center of USA

# Add points to the map
for index, row in df.iterrows():
    folium.CircleMarker(
        location=(row['latitude'], row['longitude']),
        radius=3,  # Small radius for individual points
        color='red',
        fill=True,
        fill_color='red'
    ).add_to(map)
```

```python
# Save the map to an HTML file
map.save('listings_map.html')
```
