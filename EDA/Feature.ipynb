```python
import pandas as pd
import numpy as np

# Sample hotel data
data = {
    'id': [109, 2708, 2732, 2864, 63416],
    'latitude': [33.98301, 34.09625, 34.00440, 33.87490, 34.11344],
    'longitude': [-118.38607, -118.34605, -118.48095, -118.11515, -118.32399]
}
hotels_df = pd.DataFrame(data)

# Define POIs
pois = {
    'downtown': (34.0522, -118.2437)
}

# Haversine formula to calculate distance between two points
def haversine(lon1, lat1, lon2, lat2):
    lon1, lat1, lon2, lat2 = map(np.radians, [lon1, lat1, lon2, lat2])
    dlon = lon2 - lon1
    dlat = lat2 - lat1
    a = np.sin(dlat/2)**2 + np.cos(lat1) * np.cos(lat2) * np.sin(dlon/2)**2
    c = 2 * np.arcsin(np.sqrt(a))
    km = 6371 * c  # 6371 is the radius of the Earth in kilometers
    return km

# Calculate distance to downtown for each hotel
poi = pois['downtown']
hotels_df['distance_to_downtown_km'] = hotels_df.apply(lambda x: haversine(poi[1], poi[0], x['longitude'], x['latitude']), axis=1)

# Simple rating system: inverse of distance (for demonstration)
# You can adjust the scoring based on your criteria
max_distance = hotels_df['distance_to_downtown_km'].max()
hotels_df['location_rating'] = 1 - (hotels_df['distance_to_downtown_km'] / max_distance)

print(hotels_df[['id', 'distance_to_downtown_km', 'location_rating']])
```

```python
import requests
import pandas as pd

# Example DataFrame
location_df = pd.DataFrame({
    'id': [109, 2708, 2732, 2864, 63416],
    'latitude': [33.98301, 34.09625, 34.00440, 33.87490, 34.11344],
    'longitude': [-118.38607, -118.34605, -118.48095, -118.11515, -118.32399]
})

# Function to fetch nearby POIs using Google Places API
def fetch_pois(lat, lng, api_key, radius=500, poi_type='restaurant'):
    base_url = "https://maps.googleapis.com/maps/api/place/nearbysearch/json"
    params = {
        'location': f'{lat},{lng}',
        'radius': radius,
        'type': poi_type,
        'key': api_key
    }
    response = requests.get(base_url, params=params)
    if response.status_code == 200:
        return len(response.json().get('results', []))
    else:
        print(f"Error fetching POIs for {lat}, {lng}")
        return 0

# Your Google API key
# api_key = 'AIzaSyAoGpXZZjkBDD3oiSdg0FYMaVQngH0yw6o'

# Example: Count restaurants within 500m radius for each location
location_df['restaurant_count'] = location_df.apply(lambda row: fetch_pois(row['latitude'], row['longitude'], api_key), axis=1)

# Now, calculate scores based on the counts (this part is very simplified)
max_count = location_df['restaurant_count'].max()
min_count = location_df['restaurant_count'].min()

# Normalize and scale scores to 1-5
location_df['restaurant_score'] = location_df['restaurant_count'].apply(lambda x: 1 + (4 * (x - min_count) / (max_count - min_count)) if max_count > min_count else 3)  # Default to middle score if all counts are equal

print(location_df[['id', 'restaurant_count', 'restaurant_score']])
```

```python
import numpy as np
# Step 1: Handle 'half-bath' cases by replacing them with "0.5"
df['bathrooms_text'] = df['bathrooms_text'].replace(['Private half-bath', 'Half-bath','Shared half-bath'], '0.5 bath')

# Step 2: Filter rows where both bathrooms and bathrooms_text are null
df_filtered = df.dropna(subset=['bathrooms', 'bathrooms_text'], how='all')

# Step 3: Extract numeric value from 'bathrooms_text'
# This regex pattern looks for a sequence of digits possibly followed by a dot and more digits
df_filtered['bathrooms_text_numeric'] = df_filtered['bathrooms_text'].str.extract(r'(\d+\.?\d*)').astype(float)

# Step 4: Combine 'bathrooms' and 'bathrooms_text_numeric' into one column
# np.where(condition, [x, y]) is used here: if condition is true it takes x, otherwise y.
df_filtered['bathrooms_combined'] = np.where(df_filtered['bathrooms'].notnull(), df_filtered['bathrooms'], df_filtered['bathrooms_text_numeric'])

df_filtered.head()
```

property_category has the following values:

1, apartment if property_type is Apartment, Serviced apartment, Loft.

2, hotel if property_type is Bed & Breakfast, Boutique hotel, Hostel.

3, condo if property_type is Townhouse, Condominium.

4, house if property_type is Bungalow, House.

5, other, otherwise

```python
# Python adaptation for categorization
def categorize_property(property_type):
    property_type = property_type.lower()
    if any(keyword in property_type for keyword in ["apartment", "loft", "serviced apartment"]):
        return "apartment"
    elif any(keyword in property_type for keyword in ["bed & breakfast", "boutique hotel", "hostel", "hotel", "resort", "aparthotel", "bed and breakfast"]):
        return "hotel"
    elif any(keyword in property_type for keyword in ["townhouse", "condominium", "condo"]):
        return "condo"
    elif any(keyword in property_type for keyword in ["bungalow", "house", "villa", "cottage", "chalet", "in-law", "guesthouse", "cabin"]):
        return "house"
    else:
        return "other"

# Apply the function to the DataFrame
df_filtered['property_category'] = df_filtered['property_type'].apply(categorize_property)

# To check the distribution of the new categories
category_distribution = df_filtered['property_category'].value_counts()
print(category_distribution)
```

```python
def categorize_prices(group):
    labels = ['Low', 'Medium', 'High', 'Very High']
    return pd.qcut(group, q=4, labels=labels, duplicates='drop')

df_filtered['price_category'] = df_filtered.groupby(['state', 'city'])['price'].transform(categorize_prices)

print(df_filtered[['state', 'city', 'price', 'price_category']])
```

```python
import gradio as gr

# Assuming df_filtered is your DataFrame and is already prepared

def recommend_hotels(property_category, room_type, accommodates, bathrooms_combined, beds, price_category, number_of_reviews, review_scores_rating, state, city):
    # Filter DataFrame based on user preferences
    recommendations = df_filtered[
        (df_filtered['property_category'] == property_category) &
        (df_filtered['room_type'] == room_type) &
        (df_filtered['accommodates'] >= accommodates) &
        (df_filtered['bathrooms_combined'] >= bathrooms_combined) &
        (df_filtered['beds'] >= beds) &
        (df_filtered['price_category'] == price_category) &
        (df_filtered['number_of_reviews'] >= number_of_reviews) &
        (df_filtered['review_scores_rating'] >= review_scores_rating) &
        (df_filtered['state'] == state) &
        (df_filtered['city'] == city)
    ]

    # Sort by review_scores_rating, then by number_of_reviews to get top recommendations
    recommendations = recommendations.sort_values(by=['review_scores_rating', 'number_of_reviews'], ascending=False)

    # Return top 5 recommendations
    return recommendations.head(5).to_dict('records')

# Gradio interface setup
iface = gr.Interface(
    fn=recommend_hotels,
    inputs=[
        gr.Dropdown(choices=df_filtered['property_category'].unique(), label="Property Category"),
        gr.Dropdown(choices=df_filtered['room_type'].unique(), label="Room Type"),
        gr.Slider(minimum=1, maximum=df_filtered['accommodates'].max(), label="Accommodates"),
        gr.Slider(minimum=1, maximum=df_filtered['bathrooms_combined'].max(), label="Bathrooms"),
        gr.Slider(minimum=1, maximum=df_filtered['beds'].max(), label="Beds"),
        gr.Dropdown(choices=['Low', 'Medium', 'High', 'Very High'], label="Price Category"),
        gr.Number(label="Minimum Number of Reviews", value=0),
        gr.Slider(minimum=0.0, maximum=5.0, step=0.1, label="Minimum Review Score Rating"),
        gr.Dropdown(choices=df_filtered['state'].unique(), label="State"),
        gr.Dropdown(choices=df_filtered['city'].unique(), label="City")
    ],
    outputs=gr.JSON(label="Top 5 Recommendations"),
    title="Hotel Recommendation System",
    description="Select your preferences to get the top 5 hotel recommendations."
)

# Launch the Gradio interface
iface.launch()
```
