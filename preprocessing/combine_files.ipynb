```python
import pandas as pd
import csv
```

```python
file_subnames = {
    'LA_listings': ['CA', 'Los Angeles'],
    'San_Diego_listings': ['CA', 'San Diego'],
    'San_Francisco_listings': ['CA', 'San Francisco'],
    'San_Mateo_listings': ['CA', 'San Mateo County'],
    'Santa_Clara_listings': ['CA', 'Santa Clara County'],
    'Santa_Cruz_listings': ['CA', 'Santa Cruz County'],
    'Denver_listings': ['CO', 'Denver'],
    'Washington_listings': ['DC', 'Washington D.C.'],
    'Hawaii_listings': ['HI', 'Hawaii'],
    'Chicago_listings': ['IL', 'Chicago'],
    'Boston_listings': ['MA', 'Boston'],
    'Cambridge_listings': ['MA', 'Cambridge'],
    'New_York_listings': ['NY', 'New York City'],
    'Rochester_listings': ['NY', 'Rochester'],
    'Columbus_listings': ['OH', 'Columbus'],
    'Salem_listings': ['OR', 'Salem'],
    'Nashville_listings': ['TN', 'Nashville'],
    'Austin_listings': ['TX', 'Austin'],
    'Seattle_listings': ['WA', 'Seattle']
}

# Initialize DataFrame to avoid appending to an undefined variable
airbnb_usa = pd.DataFrame()

# Iterate over each file in the dictionary
for file, location_info in file_subnames.items():
    # Create a temporary DataFrame to store values
    temp_df = pd.read_csv('/content/drive/MyDrive/DS5460 - Big Data/data/listing/' + file + '.csv')
    temp_df['state'] = location_info[0]
    temp_df['city'] = location_info[1]

    # Append to the master DataFrame
    airbnb_usa = pd.concat([airbnb_usa, temp_df], ignore_index=True)

# Specify the path where you want to save the combined dataset
save_path = '/content/drive/MyDrive/DS5460 - Big Data/data/Master CSVs/master_listings.csv'

# Save the combined dataset to the specified path
airbnb_usa.to_csv(save_path, index=False, encoding='utf-8', quoting=csv.QUOTE_NONNUMERIC)

print(f"Dataset saved to {save_path}")
```

```python
# use batches to process the Calendar files, because the ram reaches out of the availability, and session will crash if not.
file_subnames = {
    'LA_calendar': ['CA', 'Los Angeles'],
    'San_Diego_calendar': ['CA', 'San Diego'],
    'San_Francisco_calendar': ['CA', 'San Francisco'],
    'San_Mateo_calendar': ['CA', 'San Mateo County'],
    'Santa_Clara_calendar': ['CA', 'Santa Clara County'],
    'Santa_Cruz_calendar': ['CA', 'Santa Cruz County'],
    'Denver_calendar': ['CO', 'Denver'],
    'Washington_calendar': ['DC', 'Washington D.C.'],
    'Hawaii_calendar': ['HI', 'Hawaii'],
    'Chicago_calendar': ['IL', 'Chicago'],
    'Boston_calendar': ['MA', 'Boston'],
    'Cambridge_calendar': ['MA', 'Cambridge'],
    'New_York_calendar': ['NY', 'New York City'],
    'Rochester_calendar': ['NY', 'Rochester'],
    'Columbus_calendar': ['OH', 'Columbus'],
    'Salem_calendar': ['OR', 'Salem'],
    'Nashville_calendar': ['TN', 'Nashville'],
    'Austin_calendar': ['TX', 'Austin'],
    'Seattle_calendar': ['WA', 'Seattle']
}
# Function to process a batch of files and return a combined DataFrame
def process_batch(file_batch):
    batch_df = pd.DataFrame()
    for file, location_info in file_batch.items():
        temp_df = pd.read_csv(f'/content/drive/MyDrive/DS5460 - Big Data/data/calendar/{file}.csv')
        temp_df['state'] = location_info[0]
        temp_df['city'] = location_info[1]
        batch_df = pd.concat([batch_df, temp_df], ignore_index=True)
    return batch_df

# Your original file_subnames split into batches (example with two batches here)
batches = [
    {k: file_subnames[k] for k in list(file_subnames)[:5]},  # First 5 files
    {k: file_subnames[k] for k in list(file_subnames)[5:10]},# Next 5 files
    {k: file_subnames[k] for k in list(file_subnames)[10:15]},
    {k: file_subnames[k] for k in list(file_subnames)[15:19]}
]

# Initialize an empty DataFrame for the combined data
# airbnb_usa = pd.DataFrame()

# Process each batch and combine
# for batch in batches:
#     batch_df = process_batch(batch)
#     # Optionally, save each batch to disk to reduce memory usage
#     batch_df.to_csv(f'/content/drive/MyDrive/temp_batch_{batches.index(batch)}.csv', index=False, encoding='utf-8', quoting=csv.QUOTE_NONNUMERIC)
```

```python
for batch in batches:
    batch_df = process_batch(batch)
    # Optionally, save each batch to disk to reduce memory usage
    batch_df.to_csv(f'/content/drive/MyDrive/temp_batch_{batches.index(batch)}.csv', index=False, encoding='utf-8', quoting=csv.QUOTE_NONNUMERIC)
    # Read back the saved batch DataFrame (if saved to disk) or directly concatenate
    temp_batch_df = pd.read_csv(f'/content/drive/MyDrive/temp_batch_{batches.index(batch)}.csv')
    airbnb_usa = pd.concat([airbnb_usa, temp_batch_df], ignore_index=True)

# Finally, save the combined dataset
save_path = '/content/drive/MyDrive/DS5460 - Big Data/data/Master CSVs/master_calendar.csv'
airbnb_usa.to_csv(save_path, index=False, encoding='utf-8', quoting=csv.QUOTE_NONNUMERIC)

print(f"Dataset saved to {save_path}")
```
